```python
import json
import os

memo_path = "outputs/accounts/account_1/v1/account_memo.json"

with open(memo_path) as f:
    memo = json.load(f)

agent = {
    "agent_name": memo["company_name"] + " Voice Agent",
    "version": "v1",
    "voice_style": "professional",

    "system_prompt": f"""
You are an AI voice assistant for {memo['company_name']}.

BUSINESS HOURS FLOW
1. Greet the caller.
2. Ask the purpose of the call.
3. Collect caller name and phone number.
4. Transfer to service staff.
5. If transfer fails, apologize and say someone will call back.
6. Ask if they need anything else.
7. Close politely.

AFTER HOURS FLOW
1. Greet the caller.
2. Ask purpose of the call.
3. Ask if it is an emergency.
4. If emergency, collect name, phone number, and service address immediately.
5. Attempt transfer to on-call technician.
6. If transfer fails, apologize and assure quick follow-up.
7. If non-emergency, collect request details and confirm follow-up during business hours.
8. Ask if they need anything else.
9. Close the call politely.
""",

    "call_transfer_protocol": {
        "method": "phone_transfer",
        "timeout_seconds": 60,
        "retry_attempts": 1
    },

    "fallback_protocol": "If transfer fails, apologize and inform caller that the team will follow up soon."
}

agent_path = "outputs/accounts/account_1/v1/agent_spec.json"

with open(agent_path,"w") as f:
    json.dump(agent,f,indent=4)

print("Retell agent spec created")
```

```python
import re
import json
import os

INPUT_FILE = "inputs/demo_calls/account_1.txt"
OUTPUT_DIR = "outputs/accounts/account_1/v1"

with open(INPUT_FILE) as f:
    text = f.read()

# extract fields using regex
company = re.search(r"G&M Pressure washing", text, re.IGNORECASE)
phone = re.search(r"\d{3}-\d{3}-\d{4}", text)
email = re.search(r"\S+@\S+", text)

company_name = company.group(0) if company else "unknown"
phone_number = phone.group(0) if phone else "unknown"
email_id = email.group(0) if email else "unknown"

memo = {
    "account_id": "account_1",
    "company_name": company_name,
    "contact_phone": phone_number,
    "contact_email": email_id,
    "business_hours": {
        "days": "unknown",
        "start": "unknown",
        "end": "unknown",
        "timezone": "unknown"
    },
    "services_supported": ["pressure washing"],
    "questions_or_unknowns": [
        "business hours",
        "routing rules"
    ],
    "notes": "Automatically extracted from transcript"
}

with open(f"{OUTPUT_DIR}/account_memo_auto.json","w") as f:
    json.dump(memo,f,indent=4)

print("Automatic extraction completed")
```
